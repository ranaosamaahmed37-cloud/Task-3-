```python
# استيراد المكتبات الأساسية
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns

# قراءة البيانات مباشرة من الإنترنت
url = "https://gist.githubusercontent.com/pravalliyaram/5c05f43d2351249927b8a3f3cc3e5ecf/raw/Mall_Customers.csv"
data = pd.read_csv(url)

# عرض أول 5 صفوف للتأكد
print("عرض أول 5 صفوف من البيانات:")
print(data.head())

# اختيار الأعمدة اللي هنستخدمها (مثلاً Annual Income و Spending Score)
X = data.iloc[:, [3, 4]].values   # العمودين رقم 3 و 4

# استخدام طريقة Elbow لاختيار عدد الكلاسترز المناسب
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title("طريقة Elbow")
plt.xlabel("عدد الكلاسترز")
plt.ylabel("WCSS")
plt.show()

# تطبيق KMeans بعدد 5 كلاسترز (مثال شائع)
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# رسم النتائج
plt.figure(figsize=(8,6))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=100, c='magenta', label='Cluster 5')

# رسم السنتر بتاع الكلاسترز
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=300, c='yellow', marker='*', label='Centroids')

plt.title("KMeans Clustering")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage

# تحميل البيانات
data = pd.read_csv("Mall_Customers.csv")
data.head()
```

```python
data.info()
data.describe()
data.isnull().sum()
```

```python
X = data[["Annual Income (k$)", "Spending Score (1-100)"]]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker="o")
plt.xlabel("عدد الكلسترز")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
clusters_kmeans = kmeans.fit_predict(X_scaled)

plt.scatter(X_scaled[:,0], X_scaled[:,1], c=clusters_kmeans, cmap="rainbow")
plt.title("KMeans Clustering")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters_dbscan = dbscan.fit_predict(X_scaled)

plt.scatter(X_scaled[:,0], X_scaled[:,1], c=clusters_dbscan, cmap="rainbow")
plt.title("DBSCAN Clustering")
plt.show()
```

```python
linked = linkage(X_scaled, method='ward')
plt.figure(figsize=(10, 6))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=False)
plt.title("Dendrogram")
plt.show()
```

```python
agg = AgglomerativeClustering(n_clusters=5, linkage="ward")
clusters_agg = agg.fit_predict(X_scaled)

plt.scatter(X_scaled[:,0], X_scaled[:,1], c=clusters_agg, cmap="rainbow")
plt.title("Agglomerative Clustering")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=5, min_samples=5)  # ممكن تغيّري eps و min_samples حسب البيانات
clusters_dbscan = dbscan.fit_predict(X_scaled)

# check number of clusters
n_clusters = len(set(clusters_dbscan)) - (1 if -1 in clusters_dbscan else 0)
print("DBSCAN clusters:", n_clusters)

if n_clusters > 1:
    print("Silhouette Score (DBSCAN):", silhouette_score(X_scaled, clusters_dbscan))
else:
    print("DBSCAN produced less than 2 clusters, silhouette score not applicable.")
```
