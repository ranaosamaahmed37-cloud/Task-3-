```python
# 1) استيراد المكتبة
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 2) رسم بسيط
plt.plot([1,2,3,4,5])
plt.show()
```

```python
# 3) عمل Lists ورسمها
X = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(X, y)
plt.show()
```

```python
# 4) طريقة الـ OO
fig, ax = plt.subplots()
ax.plot(X, y)
plt.show()
```

```python
# 5) إضافة عنوان ومحاور وحفظ الشكل
fig, ax = plt.subplots()
ax.plot(X, y)
ax.set_title("Simple Line Plot")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
fig.savefig("line_plot.png")
plt.show()
```

```python
import numpy as np
```

```python
# 6) أرقام من 0 لـ 100
X = np.linspace(0, 100, 100)
```

```python
# 7) X مقابل X^2
fig, ax = plt.subplots()
ax.plot(X, X**2)
plt.show()
```

```python
# 8) Scatter plot لـ exp
plt.scatter(X, np.exp(X))
plt.show()
```

```python
# 9) Scatter plot لـ sin
plt.scatter(X, np.sin(X))
plt.show()
```

```python
# 10) Dictionary للأكل
foods = {"Pizza": 50, "Burger": 35, "Pasta": 45}
```

```python
# 11) Bar plot
plt.bar(foods.keys(), foods.values())
plt.title("Favourite Foods")
plt.xlabel("Food")
plt.ylabel("Price")
plt.show()
```

```python
# 12) Horizontal bar plot
plt.barh(list(foods.keys()), list(foods.values()))
plt.show()
```

```python
# 13) Normal distribution
X = np.random.randn(1000)
plt.hist(X)
plt.show()
```

```python
# 14) Random uniform distribution
X = np.random.rand(1000)
plt.hist(X)
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 5))

# Plot 1
ax1.plot(X, X/2)

# Plot 2
ax2.scatter(np.random.rand(10), np.random.rand(10))

# Plot 3
ax3.bar(foods.keys(), foods.values())

# Plot 4
ax4.hist(np.random.randn(1000))

plt.show()
```
